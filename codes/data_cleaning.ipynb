```python
# add column 'path' recording the trajectory
for i in range(df.shape[0]):
    occ_codes = [df.iloc[i,22]]
    for j in range(23,58):
        try:
            tmp = int(float(df.iloc[i,j]))
        except:
            tmp = df.iloc[i,j]

        if tmp!=occ_codes[-1]:
            occ_codes.append(tmp)

    df.loc[i,"path"]=str(occ_codes)[1:-1]

df.drop(df.columns[[*range(22,58)]], axis=1, inplace=True)
```

## 2nd step: Clean values

```python
df2 = pd.read_csv("../data/processed/drop_columns.csv")
df2.head(3)
```

#### column 'Education at start' and 'Education at end'

```python
df2["Education at start"] = df2["Education at start"].str[0:-2]
df2["Education at end"] = df2["Education at end"].str[0:-3]
```

```python
df2["Education at start"].unique()
```

```python
df2["Education at end"].unique()
```

```python
for i in range(df2.shape[0]):
    for j in ["Education at start", "Education at end"]:
        if df2.loc[i,j]=='educNoDeg':
            df2.loc[i,j]='no degrees'
        elif df2.loc[i,j]=='educHS':
            df2.loc[i,j]='HS diploma or equivalent'
        elif df2.loc[i,j]=='educAA':
            df2.loc[i,j]='associate degree'
        elif df2.loc[i,j]=='educBA':
            df2.loc[i,j]='bachelor\'s degree'
```

#### column 'Starting occupational cluster' and 'Alternate starting occupational cluster'

```python
df2["Starting occupational cluster"].unique()
```

```python
df2["Alternate starting occupational cluster"].unique()
```

```python
comparison_table = {"sect_Other": 'All other', "sect_Construct": "Construction", "sect_EdLegSoc": "Educ/Legal/Social Services", 
                    "sect_EngnrSci": "Engineer/Sci/Arch", "sect_Health": "Healthcare", "sect_IT": "IT",
                    "sect_MaintRepr": "Maintance/Repair", "sect_Mgmtfin": "Mgmt/Biz/Financial", "sect_OfficeAdm": "Office & Admin Sppt", 
                    "sect_PerServc": "Personal Service", "sect_Production": "Production", "sect_Protective": "Protective Services",
                    "sect_Sales": 'Sales', "sect_Transport": "Transportation",
                    "sectalt_EdECE": 'Early Childhood Education', "sectalt_STEM": 'STEM', 
                    "sectalt_OtherAg": 'Agriculture', "sectalt_OtherArts": 'Arts and Entertainment'}


for i in range(df2.shape[0]):
    for j in ["Starting occupational cluster", "Alternate starting occupational cluster"]:
        if pd.isna(df2.loc[i,j]) is not True:
            df2.loc[i,j] = comparison_table[df2.loc[i,j]]
```

#### Extract edges and their weights from column 'path'

```python
edges_and_weight = {}

for i in range(df2.shape[0]):
    path = df2.loc[i,'path'].split(", ")
    for n in range(len(path)-1):
        nodes = "{},{}".format(path[n],path[n+1])
        if nodes in edges_and_weight.keys():
            edges_and_weight[nodes]+=1
        else:
            edges_and_weight[nodes]=1
```

```python
df_edges_weight = pd.DataFrame({'edge':list(edges_and_weight.keys()), 'weight':list(edges_and_weight.values())})
df_edges_weight
```

```python
df_edges_weight[["source","target"]] = df_edges_weight["edge"].str.split(",", expand=True)
df_edges_weight.drop(columns=["edge"], inplace=True)
df_edges_weight
```

```python
df_edges_weight.to_csv("../data/processed/edges_and_weight.csv",index=False)
```
